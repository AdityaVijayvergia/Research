```
import os
import numpy as np
glove_dir = '/content'
embeddings_index = {}
f = open(os.path.join(glove_dir, 'glove.6B.100d.txt'))
for line in f:
  values = line.split()
  word = values[0]
  coefs = np.asarray(values[1:], dtype='float32')
  embeddings_index[word] = coefs
f.close()
print('Found %s word vectors.' % len(embeddings_index))
```

```
word_list = list(embeddings_index.keys())
```

```
word_list[0]
```

```
word_index = {}

for i in range(len(word_list)):
    word_index[word_list[i]] = i
```

```
input_train = []
for eg in X:
    in_eg = []
    for word in eg:
        ind = word_index.get(word)
        if ind is not None:
            in_eg.append(ind)
    input_train.append(in_eg)
```

```
maxlen=50
input_train = sequence.pad_sequences(input_train, maxlen=maxlen)
```

```
input_train.shape
```

```
input_train[0]
```

```
from keras.layers import Dense, Embedding, LSTM, Flatten
from keras.models import Sequential
```

```
len(embeddings_index[word_list[0]])
```

```
no_of_words = len(word_list)
embedding_dim = len(embeddings_index[word_list[0]])

model = Sequential()
model.add(Embedding(no_of_words, embedding_dim))
model.add(LSTM(16, dropout=0.15, recurrent_dropout=0.15, return_sequences=True))
model.add(LSTM(8, dropout=0.15, recurrent_dropout=0.15))
model.add(Dense(1, activation='sigmoid'))
```

```
model.summary()
```

```
model.layers
```

```
embedding_matrix = np.zeros((no_of_words, embedding_dim))
for word, i in word_index.items():
    embedding_matrix[i] = embeddings_index.get(word)
```

```
embedding_matrix.shape
```

```
model.layers[0].set_weights([embedding_matrix])
model.layers[0].trainable = False
```

```
model.summary()
```

```
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
```

```
history = model.fit(input_train, y, epochs=25, batch_size=32, validation_split=0.15)
```
